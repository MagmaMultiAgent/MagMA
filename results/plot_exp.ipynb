```python
import pandas as pd
import matplotlib.pyplot as plt

def plot(data_frame, y_label, title, smoothing):

    plt.figure(figsize=(10, 6))

    for i, df in enumerate(data_frame):
        smoothed_data = df['Value'].rolling(window=smoothing).mean()
        plt.plot(df['Step'], smoothed_data, label=f'Algo: {algo_names[i]}')

    plt.xlabel('Step (in 1e6)')
    plt.ylabel(y_label)
    plt.xlim(0, 3000000)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

algo_names = ['A2C', 'DQN', 'MPPO', 'PPO', 'QRDQN', 'RPPO', 'TRPO']
```

```python
file_names_rew = ['A2C/2023.08.25./A2C_1/data/ep_rew_mean.csv','DQN/2023.08.23./DQN_1/data/ep_rew_mean.csv', \
            'MPPO/2023.08.28./MPPO_1/data/ep_rew_mean.csv', 'PPO/2023.08.23./PPO_1/data/ep_rew_mean.csv', 'QRDQN/2023.08.25./QRDQN_1/data/ep_rew_mean.csv',\
            'RPPO/2023.08.24./RPPO_1/data/ep_rew_mean.csv', 'TRPO/2023.08.25./TRPO_1/data/ep_rew_mean.csv']

data_frames_rew = [pd.read_csv(filename) for filename in file_names_rew]

plot(data_frames_rew, "Mean Reward", "Combined Plot of Mean Rewards", 1)
```
